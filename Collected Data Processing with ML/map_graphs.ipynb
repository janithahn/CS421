```python
fig, ax = plt.subplots(figsize=(10,10))
for key, grp in gr:
    if key == 'TreeMap':
        pass
    ax.set_ylabel('run-time');
    ax.set_xlabel('NumberOfElements');
    ax.plot(grp['NumberOfIntegers'], grp['containsValue()'], label=key + "_search()")

ax.legend()
plt.show()
```

```python
df_memory = pd.read_csv('./data/map_memory_20steps.csv')
```

```python
df_memory.head(10)
```

```python
gm = df_memory.groupby('DataStructure')
```

```python
map_groups_memory = [gm.get_group(x) for x in gm.groups]
```

```python
map_groups_memory[0]
```

```python
fig, ax = plt.subplots(figsize=(10,10))
for key, grp in gm:
    ax.set_ylabel('memory');
    ax.set_xlabel('NumberOfElements');
    ax.plot(grp['NumberOfIntegers'], grp['put()'], label=key + "_put()")

ax.legend()
plt.show()
```
